```python
import requests
import networkx as nx
from tqdm.notebook import tqdm

url = 'https://api.binance.com/api/v3/exchangeInfo?symbolStatus=TRADING&showPermissionSets=false'
G = nx.Graph()

# 124129610835878713695549635770708419868047758589952 ~ 1.2e+50 - оценка количества циклов в графе на 405 вершинах и 1329 ребрах
```

```python
r = requests.get(url).json()
for sym in r["symbols"]:
    G.add_node(sym["baseAsset"])
    G.add_node(sym["quoteAsset"])
    G.add_edge(sym["baseAsset"], sym["quoteAsset"])
```

```python
def print_graph(G):
    pos = nx.circular_layout(G)
    options = {
        'node_color': 'blue',
        'node_size': 1,
        'width': 1,
        'edge_color': 'black'
    }

    print(G)
    nx.draw(G, pos, **options)
```

```python
print_graph(G)
```

```python
nodes = list(G.nodes())
line = [nodes[0]]
iterators = dict([(i, G.neighbors(i)) for i in nodes])
cnters = {nodes[0]: 0}
cycles = {}

max_cycles = 20_000_000
pbar = tqdm(total=max_cycles // 1_000_000)
cycle_num = 0

while len(line) != 0 and cycle_num < max_cycles - 1:  # while len(line) != 0:
    node = line[-1]

    while True:
        try:
            next_node = next(iterators[node])
            if next_node not in cnters:
                cnters[next_node] = len(line)
                line += [next_node]
                iterators[next_node] = G.neighbors(next_node)
                break
            elif len(line) - cnters[next_node] >= 3:
                way = tuple(line[cnters[next_node]:])
                h_way = hash(tuple(sorted(way)))
                if h_way not in cycles:
                    cycles[h_way] = way
                    cycle_num += 1

                    if cycle_num % 1_000_000 == 0:
                        pbar.update()
        except StopIteration:
            line.pop()
            cnters.pop(node)
            break
```

```python
def ccls(start_node, iterators_of_nodes, cycles, len_of_cycle):
    line = [start_node]
    iterators = iterators_of_nodes.copy()
    cnters = {start_node: 0}

    while len(line) != 0:
        node = line[-1]

        while True:
            try:
                next_node = next(iterators[node])
                if next_node not in cnters:
                    if len(line) < len_of_cycle:
                        cnters[next_node] = len(line)
                        line += [next_node]
                        iterators[next_node] = iterators_of_nodes[next_node]
                        break

                elif len(line) - cnters[next_node] == len_of_cycle:
                    way = tuple(line[cnters[next_node]:])
                    h_way = hash(tuple(sorted(way)))
                    if h_way not in cycles:
                        cycles[h_way] = way

            except StopIteration:
                line.pop()
                cnters.pop(node)
                break

    return cycles
```

```python
iterators_of_nodes = dict([(i, G.neighbors(i)) for i in G.nodes()])
cycles = {}

for node in list(G.nodes()):
    cycles = ccls(node, iterators_of_nodes, cycles.copy(), 3)
```

```python
tric = []
for i in G.nodes():
    for j in G.nodes():
        for k in G.nodes():
            if G.has_edge(i, j) and G.has_edge(j, k) and G.has_edge(i, k):
                st = sorted([i, j, k])
                if st not in tric:
                    tric += [st]
print(len(tric))
```
